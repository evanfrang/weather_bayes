```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
from meteostat import Daily, Stations

```

```python
location_name = "Los Angeles"
start = datetime(2018, 1, 1)
end = datetime(2023, 12, 31)

stations = Stations()
stations = stations.nearby(34.0522, -118.2437)
station = stations.fetch(1)

print("Using station:", station.iloc[0]['name'])

data = Daily(station.index[0], start, end)
data = data.fetch()

data = data[["tavg"]]  # 'tavg' = average daily temperature in °C

data = data.dropna()

plt.figure(figsize=(10, 5))
plt.plot(data.index, data["tavg"], label="Daily Avg Temp (°C)")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title(f"Daily Average Temperature - {location_name}")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

data.to_csv("la_temperature.csv")
print("Saved data to la_temperature.csv")
```
